```julia
h = 0.1
n = 100
```

```julia
# First-order ODE
# Load necessary packages
using Plots
using Statistics

# Define the exact solution
function exact_solution(x)
    return x*(x+1)
end

# Define the derivative function
function deriv_func(x, y)
    return 2*x+1
end

# Define the Forward Euler method
function forward_euler(deriv_func, x0, y0, h, n)
    t0 = time()
    x = x0
    y = y0
    y_list = [y0]
    for i in 1:n
        y = y + h*deriv_func(x, y)
        x = x + h
        push!(y_list, y)
    end
    t1 = time()
    return convert(Array{Float64,1}, y_list), (t1-t0)*1000
end

# Define the Backward Euler method
function backward_euler(deriv_func, x0, y0, h, n)
    t0 = time()
    x = x0
    y = y0
    y_list = [y0]
    for i in 1:n
        y_next = y + h*deriv_func(x, y)
        y = y + h*deriv_func(x+h, y_next)
        x = x + h
        push!(y_list, y)
    end
    t1 = time()
    return convert(Array{Float64,1}, y_list), (t1-t0)*1000
end

# Set the initial conditions and parameters
x0 = 0
y0 = 0.0

# Calculate the exact solution
x = range(x0, stop=x0+h*n, length=n+1)
y_exact = exact_solution.(x)

# Calculate the Forward Euler solution and time
y_forward, t_forward = forward_euler(deriv_func, x0, y0, h, n)

# Calculate the Backward Euler solution and time
y_backward, t_backward = backward_euler(deriv_func, x0, y0, h, n)

# Calculate the tolerance values
tol_forward = abs.(y_forward - y_exact)
tol_backward = abs.(y_backward - y_exact)

# Plot the results
plot(x, y_exact, label="Exact solution")
xlabel!("x")
ylabel!("y")
title!("Comparison of ODE solutions")
plot!(titlefont=font(12))
display(plot!(x, y_forward, label="Forward Euler"))
display(plot!(x, y_backward, label="Backward Euler"))

plot(tol_forward, label="Forward Euler")
plot!(tol_backward, label="Backward Euler")
xlabel!("x")
ylabel!("Error")
title!("Error of ODE solutions")
plot!(titlefont=font(12))
display(plot!(tol_backward, label="Backward Euler"))

println("n = $n, h = $h, y(x) = x(x+1)")
println("Time taken by Forward Euler method: $t_forward")
println("Time taken by Backward Euler method: $t_backward")

APE_forward = mean(abs.(tol_forward./y_exact)[2:end])*100
APE_backward = mean(abs.(tol_backward./y_exact)[2:end])*100

# Print the results
println("Average percentage error of Forward Euler method: $APE_forward")
println("Average percentage error of Backward Euler method: $APE_backward")
```
